## metrics of inceptionv2 with arcface loss

```python
metrics = {}
with open("/kaggle/input/metric-arc-incep/metrics_arcloss_inception.txt", "r") as f:
    for line in f:
        key, value = line.split("=")
        metrics[key.strip()] = eval(value.strip())  
```

```python
train_losses = metrics["train_loss"]
accs = metrics["accuracy"]
rocs = metrics["roc_auc"]
tfs = metrics["tar_far_1e3"]
```

```python
import matplotlib.pyplot as plt

epochs = range(len(train_losses))

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(epochs, train_losses)
plt.title("Train Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True, linestyle='--', alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(epochs, accs)
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("ACC")
plt.grid(True, linestyle='--', alpha=0.3)

plt.subplot(2, 2, 3)
plt.plot(epochs, rocs)
plt.title("ROC AUC")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.grid(True, linestyle='--', alpha=0.3)

plt.subplot(2, 2, 4)
plt.plot(epochs, tfs)
plt.title("TAR@FAR")
plt.xlabel("Epoch")
plt.ylabel("TAR")
plt.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
